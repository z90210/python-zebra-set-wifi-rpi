```python
import subprocess, time, re
from termcolor import colored
from netaddr import IPAddress
import ipaddress

from passlib.utils import pbkdf2
import binascii

from prettytable import PrettyTable

print("*** WiFi斑马打印机接入配置 ***")
print("")
print("请确认下列准备事项：")
print("1.本电脑右上角之Wifi图标己接入")
print("2.具有WiFi功能之斑马打印机，USB线连接好本电脑与打印机")
print("3.打印机已开机，并在待机的状态")
print("4.已准备好WiFi之密码")

input('请按Enter开始')

print("扫描Wifi状态...")

try:
    wireless=subprocess.run("sudo iwlist wlan0 scan | grep 'ESSID\|Cell\|Frequency\|Authentication\|Quality\|Channel\|Encryption'", shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)

except AppError as error:
    #logger.error(error)
    print("error")
    raise  

if wireless.stderr:  #如果subprocess.stderr有内容，表示发生异常，中止程序
    print("扫描Wifi异常，中止程序")
    print(wireless.stderr)
    raise #中止程序

print("扫描完成")
# print(wireless.stdout) #列出扫描结果
print("Goal")
```
